```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
df_data = pd.read_pickle('df_data-pkl')
```

```python
# Number of jobs per Job title 
df_plot = df_data['job_title_short'].value_counts().to_frame()

sns.set_theme(style='ticks')
sns.barplot(data=df_plot, x='count', y='job_title_short', hue='count', palette='dark:b_r', legend=False)
sns.despine()
plt.title('Number of Jobs per Job Title')
plt.xlabel('Number of Jobs')
plt.ylabel('')
plt.show()
```

```python
# Number of jobs per country 

df_plot = df_data['job_country'].value_counts().head(20).to_frame()

sns.set_theme(style='ticks')
sns.barplot(data=df_plot, x='count', y='job_country', hue='count', palette='dark:b_r', legend=False)
sns.despine()
plt.title('Number of Jobs per Country')
plt.xlabel('Number of Jobs')
plt.ylabel('')
plt.show()
```

```python
# Job opportunities 
dict_column = {
    'job_work_from_home': 'Work from Home Offered',
    'job_no_degree_mention': 'Degree Requirement',
    'job_health_insurance': 'Health Insurance Offered'
}

fig, ax = plt.subplots(1, 3, figsize=(11, 3.5))

for i, (column, title) in enumerate(dict_column.items()):
    ax[i].pie(df_data[column].value_counts(), labels=['False', 'True'], autopct='%1.1f%%', startangle=90)
    ax[i].set_title(title)

plt.show()
```

```python
# Filter for Portugal Jobs 
df_PT = df_data[(df_data['job_country'] == 'Portugal')]
```

```python
# Portugal Locations

df_plot = df_PT['job_location'].value_counts().head(10).to_frame()

sns.set_theme(style='ticks')
sns.barplot(data=df_plot, x='count', y='job_location', hue='count', palette='dark:b_r', legend=False)
sns.despine()
plt.title('Counts of Job Locations for Data Analyst in PT')
plt.xlabel('Number of Jobs')
plt.ylabel('')
plt.show()
```

```python
# Job opportunities in Portugal
import matplotlib.pyplot as plt

dict_column = {
    'job_work_from_home': 'Work from Home Offered',
    'job_no_degree_mention': 'Degree Requirement',
    'job_health_insurance': 'Health Insurance Offered'
}

fig, ax = plt.subplots(1, 3, figsize=(11, 3.5))

for i, (column, title) in enumerate(dict_column.items()):
    # Get the value counts and labels dynamically
    counts = df_PT[column].value_counts()
    labels = counts.index.astype(str)  # Ensure labels are strings

    # Create pie chart
    ax[i].pie(counts, labels=labels, autopct='%1.1f%%', startangle=90)
    ax[i].set_title(title)

plt.tight_layout()  # Adjust spacing between subplots
plt.show()
```

```python
# Companies in Portugal
df_plot = df_PT['company_name'].value_counts().head(10).to_frame()

sns.set_theme(style='ticks')
sns.barplot(data=df_plot, x='count', y='company_name', hue='count', palette='dark:b_r', legend=False)
sns.despine()
plt.title('Counts of Companies for Data Analyst in PT')
plt.xlabel('Number of Jobs')
plt.ylabel('')
plt.show()
```
